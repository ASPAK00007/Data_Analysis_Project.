```python
#  we are dropping coloumn





df = df.drop(['EmployeeCount', 
              'EmployeeNumber', 
              'Over18', 
              'StandardHours', 
              'DailyRate', 
              'HourlyRate', 
              'MonthlyRate'], axis=1)
```

```python
# See unique values
print(df['Attrition'].unique())

# Check the first few rows
print(df['Attrition'].head(10))


```

```python
total_emp = len(df)

employee_left = df[df['Attrition']== 'Yes'].shape[0]

attrition_rate = (employee_left / total_emp ) * 100
print(f"Overall Attrition Rate: {attrition_rate:.2f}%")
```

```python
# Step 4: Attrition by Department
attrition_by_dept  = df.groupby('Department')['Attrition'].value_counts().unstack().fillna(0)

attrition_by_dept ['Attrition %'] = (att_by_dept['Yes'] / (att_by_dept['Yes'] + att_by_dept['No'])) * 100

attrition_by_dept 
```

```python
# Attrition by Job Role & Salary

attrition_by_role  = df.groupby('JobRole').agg(
    total_emp = ('Attrition' , 'count'),
    employee_left =  ('Attrition' , lambda x: (x=='Yes').sum()),
    avg_salary = ('MonthlyIncome' , 'mean')
)

attrition_by_role['Attrition %'] = (attrition_by_role['employee_left'] / (attrition_by_role['total_emp'] )) * 100
attrition_by_role = attrition_by_role.sort_values('Attrition %', ascending=False)
attrition_by_role
```

```python
attrition_overtime = df.groupby(['OverTime','JobSatisfaction']).agg(
    total_emp = ('Attrition', 'count'),
    employee_left = ('Attrition', lambda x: (x=='Yes').sum())
)

attrition_overtime['Attrition %'] = (attrition_overtime['employee_left'] / (attrition_overtime['total_emp'])) * 100
attrition_overtime = attrition_overtime.sort_values('Attrition %', ascending=False)
attrition_overtime
```

```python
# Create age groups
bins = [18, 25, 30, 40, 50, 60]
labels = ['18-25','26-30','31-40','41-50','51-60']
df['AgeGroup'] = pd.cut(df['Age'], bins=bins, labels=labels)

attrition_by_age = df.groupby('AgeGroup')['Attrition'].value_counts().unstack().fillna(0)
attrition_by_age['Attrition %'] = (attrition_by_age['Yes'] / (attrition_by_age['Yes'] + attrition_by_age['No'])) * 100
attrition_by_age
```
