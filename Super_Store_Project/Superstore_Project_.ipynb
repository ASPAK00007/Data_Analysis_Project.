```python
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\mrak5\pandas.p\Superstore.csv" , encoding="latin1")
```

```python
df.columns = df.columns.str.replace(" ", "_")
print(df.columns)
```

```python
df.info()
```

```python
df.describe()
```

```python
# Convert all column names to snake_case
df.columns = df.columns.str.lower().str.replace(" ", "_")

print(df.columns.tolist())
```

```python
df.head(7)
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.dtypes
```

```python
df["order_date"] = pd.to_datetime(df['order_date'], errors='coerce')
df["ship_date"] = pd.to_datetime(df['ship_date'], errors='coerce')
```

```python
df['delivery_days'] =(df['ship_date'] - df['order_date']).dt.days
```

```python
df.info()
```

```python
df.dropna(subset=['sales','profit','discount','quantity'], inplace=True)
```

```python
df[['customer_name','product_name','category','sub-category']] = (
    df[['customer_name','product_name','category','sub-category']].fillna('Unknown')
)
```

```python
df.info()
```

```python
summary = df[['sales','profit','discount','quantity']].describe()
summary.loc['median'] = df[['sales','profit','discount','quantity']].median()
summary
```

```python
# Sales vs Profit → usually positive.
# Discount vs Profit → often negative (more discount reduces profit).

df[['sales','profit','discount']].corr()
```

```python
df.groupby("customer_name")["profit"].sum().sort_values(ascending=False).head(10)
```

```python
loss_customers = df.groupby("customer_name")["profit"].sum()
loss_customers = loss_customers[loss_customers < 0]
```

```python
df.info()
```

```python
# total per customers
order_per_customers = df.groupby("customer_name")["order_id"].nunique()
order_per_customers
```

```python
# sub-category with highest avreage discount
suc_discount = df.groupby("sub-category")["discount"].mean().sort_values(ascending=False)
suc_discount
```

```python
# product with highest profit

product_profit =df.groupby("product_name")["profit"].sum().sort_values(ascending=False)
product_profit
```

```python
product_profit.tail()
```

```python
df['order_month'] = df['order_date'].dt.to_period('M')
monthly_sales = df.groupby('order_month')['sales'].sum()
monthly_sales
```

```python
# delivery perforemance
delvery_performance =df.groupby("sub-category")["delivery_days"].mean().sort_values()
```
